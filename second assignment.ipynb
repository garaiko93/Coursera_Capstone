```python
import requests # library to handle requests
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
import random # library for random number generation
```

```python
# download data from wikipedia
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
df = data[0]
```

```python
# delete not assigned boroughs
df = df[~df.Borough.str.contains("Not assigned")]
```

```python
coord_data = pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
# join both dataframes according to the postal codes
result = df.join(coord_data.set_index('Postal Code'), on='Postal Code')
```
